```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
iris =pd.read_csv('./datasets/iris.csv')
```

```python
iris.head(5)
```

scatterplot

```python
iris['Species'].replace({"Iris-setosa":0,'Iris-versicolor':1,'Iris-virginica':2},inplace=True)
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=1,cmap='winter')
plt.xlabel('Sepal length')
plt.ylabel('petal length')
plt.colorbar()
plt.legend()
plt.show()
```

plot size

```python
plt.figure(figsize=(6.5,6))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],alpha=1,cmap='winter')
plt.xlabel('Sepal length')
plt.ylabel('petal length')
plt.colorbar()
plt.legend()

plt.savefig('sample')
```
