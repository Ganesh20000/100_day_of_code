scatterplot two on area and population

```python

n = 100

# create random dataset
np.random.seed(42)  # for reproducibility
data = {
    "city": [f"City_{i}" for i in range(1, n+1)],
    "population": np.random.randint(1, 25, n),     # in millions
    "area": np.random.randint(100, 2000, n),       # km²
    "gdp": np.random.randint(20, 400, n)           # billion USD
}

df = pd.DataFrame(data)
print(df.head())
```

```python
plt.Figure(figsize=(10,9))
plt.scatter(df['area'],df['population'],s=df['gdp'],label='GDP')

for i in range(df.shape[0]):
    plt.text(df['area'].values[i],df['population'].values[i],df['city'].values[i],fontsize=8,fontweight="bold")
plt.title("City Area vs Population", fontsize=16, fontweight="bold")
plt.xlabel("Area (sq km)", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
