scatterplot two on area and population

```python

n = 100

# create random dataset
np.random.seed(42)  # for reproducibility
data = {
    "city": [f"City_{i}" for i in range(1, n+1)],
    "population": np.random.randint(1, 25, n),     # in millions
    "area": np.random.randint(100, 2000, n),       # km
    "gdp": np.random.randint(20, 400, n)           # billion USD
}

df = pd.DataFrame(data)
print(df.head())
```

```python
plt.Figure(figsize=(10,9))
plt.scatter(df['area'],df['population'],s=df['gdp'],label='GDP')

for i in range(df.shape[0]):
    plt.text(df['area'].values[i],df['population'].values[i],df['city'].values[i],fontsize=8,fontweight="bold")
plt.title("City Area vs Population", fontsize=16, fontweight="bold")
plt.xlabel("Area (sq km)", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

surface plots

```python
x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
```

```python
xx ,yy =np.meshgrid(x,y)
xx.shape
```

```python
z=xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
y =ax.plot_surface(xx,yy,z,cmap='winter')
plt.colorbar(y)
```

```python
z= np.sin(xx)+ np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
y =ax.plot_surface(xx,yy,z,cmap='winter')
plt.colorbar(y)
```

```python
z= np.sin(xx)+ np.log(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
y =ax.plot_surface(xx,yy,z,cmap='winter')
plt.colorbar(y)
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate 1000 days
dates = pd.date_range("2020-01-01", periods=1000, freq="D")

# Random walk stock prices
company_a = 40 + np.random.randn(1000).cumsum()
company_b = 60 + np.random.randn(1000).cumsum()
company_c = 40+ np.random.randn(1000).cumsum()

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Company A": company_a.round(2),
    "Company B": company_b.round(2),
    "Company C": company_c.round(2)
})
df.shape
```
