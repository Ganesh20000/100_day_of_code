```python
plt.plot(bat['index'],bat['RG Sharma'],color='#E7180B')
plt.plot(bat['index'],bat['V Kohli'],color='#2AA63E')
plt.title('rohit sharma vs virat kohi perfomance on run vs season ')
plt.xlabel('season')
plt.ylabel('runs scoread')
```

```python
plt.plot(bat['index'],bat['RG Sharma'],color='#0A0C0F',linestyle='dashed',linewidth=3.9)
plt.plot(bat['index'],bat['V Kohli'],color='#737373',linestyle='solid',linewidth=1.6)
plt.title('rohit sharma vs virat kohi perfomance on run vs season ')
plt.xlabel('season')
plt.ylabel('runs scoread')
```

```python
plt.plot(bat['index'],bat['RG Sharma'],color="#A30000",marker='D',markersize=3.5,linestyle='dashdot',linewidth=3.5,label='rohit')
plt.plot(bat['index'],bat['V Kohli'],color="#0DFE35",marker='D',markersize=2.5,linestyle='dotted',linewidth=2.5,label='virat')
plt.title('rohit sharma vs virat kohi perfomance on run vs season ')
plt.xlabel('season')
plt.ylabel('runs scoread')
```
