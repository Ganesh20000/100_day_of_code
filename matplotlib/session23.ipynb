multiple bar plot 

```python
plt.bar(np.arange(s.shape[0])-0.2,s['2015'],width=0.2,color='#90E31C')
plt.bar(np.arange(s.shape[0]),s['2016'],width=0.2,color='#E7180B')
plt.bar(np.arange(s.shape[0])+0.2,s['2017'],width=0.2,color='#23AFA1')
plt.xticks(np.arange(s.shape[0]),s['batsman'])
plt.show()
```

```python
plt.barh(np.arange(s.shape[0])-0.2,s['2015'],height=0.2,color='#23AFA1')
plt.barh(np.arange(s.shape[0]),s['2016'],height=0.2,color='#0C0A09')
plt.barh(np.arange(s.shape[0])+0.2,s['2017'],height=0.2,color='#7512ED')
plt.yticks(np.arange(s.shape[0]),s['batsman'])
plt.show()
```

Stacked bar chart

```python
s
```

```python
plt.bar(s['batsman'],s['2017'],label='2017',color='#C81609',width=0.5,edgecolor='black')
plt.bar(s['batsman'],s['2016'],bottom=s['2017'],label='2016',color='#0E0C0B',width=0.5,edgecolor='black')
plt.bar(s['batsman'],s['2015'],bottom=(s['2016']+s['2017']),label='2015',color="#E7E7E7",width=0.5,edgecolor='black')
plt.legend()
plt.show()
```
