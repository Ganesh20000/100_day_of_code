merging and joining and concatination

```python
import pandas as pd
import numpy as np

```

```python
employee = pd.DataFrame({
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'Marketing'],
    'HireDate': ['2018-04-23', '2019-06-01', '2020-01-15', '2017-09-10', '2021-03-05'],
    'Salary': [60000, 75000, 70000, 80000, 65000]
})
```

```python
performance = pd.DataFrame({
    'EmployeeID': [101, 102, 106, 107, 104],
    'PerformanceRating': ['Excellent', 'Good', 'Average', 'Excellent', 'Poor'],
    'salary': [5000, 3000, 1500, 5500, 0],
    'LastReviewDate': ['2024-12-10', '2024-11-15', '2024-10-30', '2024-12-01', '2024-09-20'],
    'ProjectCount': [5, 3, 2, 6, 1]
})
```

```python
employee.info()
```

```python
pd.merge(employee,performance)
```

```python
pd.merge(employee,performance,on='EmployeeID',how='inner')
```

```python
pd.merge(employee,performance,on='EmployeeID',how='outer')
```

```python
pd.merge(employee,performance,on='EmployeeID',how='left')
```

```python
pd.merge(employee,performance, on=['EmployeeID'], how='right')
```

concatination fo two dataframes

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    })
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    })
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=0)
```

joining 2 dataframes

```python
df11 = pd.DataFrame({
    'name':['alice','bob','peter','karl']
}, index=[1,2,3,4])
```

```python
df12= pd.DataFrame({
    'score':[88,95,65,76]
},index=[2,3,4,5])
```

```python
df11.join(df12)
```

```python
df11.join(df12,how="outer")
```

```python
df11.join(df12,how="inner")
```

```python
df12.join(df11,how='left')
```

```python
df12.join(df11,how="right")
```
